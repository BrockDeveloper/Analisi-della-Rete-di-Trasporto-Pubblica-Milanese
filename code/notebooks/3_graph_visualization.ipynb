```python
import igraph as ig
import matplotlib.pyplot as plt
```

```python
graph_file = "../assets/static/graphs/atm_network.pkl"
G = ig.Graph.Read_Pickle(graph_file)
```

# Visualizzazione dell'intera rete

```python
y_coords = list(G.vs['latitude'])
x_coords = list(G.vs['longitude'])

plt.figure(figsize=(12, 10))

# Visualizzazione degli archi
for e in G.es:
    v1, v2 = e.tuple
    x = [x_coords[v1], x_coords[v2]]
    y = [y_coords[v1], y_coords[v2]]
    plt.plot(x, y, color="#bbbbbb", linewidth=0.8, alpha=0.5, zorder=1)

# Visualizzazione dei nodi sopra gli archi
scatter = plt.scatter(x_coords, y_coords, s=50)
plt.xlabel('Longitudine')
plt.ylabel('Latitudine')
plt.tight_layout()
plt.show()
```

# Visualizzazione di ogni linea della rete

```python
for line in range(1, 6):
    subgraph = G.subgraph([v.index for v in G.vs if line in v['lines']])
    
    y_coords = list(subgraph.vs['latitude'])
    x_coords = list(subgraph.vs['longitude'])
    
    plt.figure(figsize=(12, 10))
    
    # Visualizzazione degli archi
    for e in subgraph.es:
        v1, v2 = e.tuple
        x = [x_coords[v1], x_coords[v2]]
        y = [y_coords[v1], y_coords[v2]]
        plt.plot(x, y, color="#bbbbbb", linewidth=0.8, alpha=0.5, zorder=1)
    
    # Visualizzazione dei nodi sopra gli archi
    scatter = plt.scatter(x_coords, y_coords, s=50)
    plt.xlabel('Longitudine')
    plt.ylabel('Latitudine')
    plt.tight_layout()
    plt.show()
```
